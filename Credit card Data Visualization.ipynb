```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cust = pd.read_csv('E:/data science assignment/4. Pandas Case Study 2 - Credit Card Case Study/Customer Acqusition.csv')
repay = pd.read_csv('E:/data science assignment/4. Pandas Case Study 2 - Credit Card Case Study/Repayment.csv')
spend = pd.read_csv('E:/data science assignment/4. Pandas Case Study 2 - Credit Card Case Study/spend.csv')
cust
```

```python
repay
```

```python
fin = pd.merge(cust, spend, left_on ='Customer', right_on = 'Customer', how = 'left' )
fin
```

```python
final = pd.merge(fin,repay,left_on = 'Customer', right_on = 'Customer',how='left')
final
```

```python
#In case age is less than 18, replace it with mean of age values
final.loc[final.Age <18,'Age'] = 18
final
```

```python
#In case spend amount is more than the limit, replace it with 50% of that customer’s limit. 
#(customer’s limit provided in acquisition table is the per transaction limit on his card)
final.loc[final.Amount_x> final.Limit,'Amount_x']= final.Limit/2
final
```

```python
#Incase the repayment amount is more than the limit, replace the repayment with the limit.

final.loc[final.Amount_y > final.Limit,'Amount_y'] = final.Limit
final
```

```python
#What is the average monthly spend by customers?
print('The average spending of customer is',final.Amount_x.mean())
```

```python
#What is the average monthly repayment by customers?
print('The average repayment by the Customer is  ', final.Amount_y.mean())
```

```python
#If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? 
#(Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment 
#– Monthly spend. Interest is earned only on positive profits and not on negative amounts)


profit  = final.Amount_y - final.Amount_x
ttl_prfit = 0.029* sum(profit[profit > 0])
print('The total profit of the bank is',ttl_prfit)
```

```python
#What are the top 5 product types?
final.groupby('Type').Amount_x.sum().reset_index().sort_values('Amount_x',ascending = False).head(5)
```

```python
#Which city is having maximum spend
final.groupby('City').Amount_x.sum().reset_index().sort_values('Amount_x',ascending =False).head(1)
```

```python
#Which age group is spending more money?
final.loc[final.Age < 25, 'Age_group'] = 'Youth'
final.loc[(final.Age >= 25) & (final.Age<=60), 'Age_group'] = 'Adult'
final.loc[final.Age > 60, 'Age_group'] = 'Senior'
final.groupby('Age_group').Amount_x.sum().reset_index().sort_values('Amount_x',ascending=False).head(1)
```

```python
# Who are the top 10 customers in terms of repayment?
final.groupby('Customer').Amount_y.sum().reset_index().sort_values('Amount_y',ascending = False).head(10)
```

```python
#Calculate the city wise spend on each product on yearly basis. Also include a graphical 
representation for the same.
final['Month_x']= pd.to_datetime(final.Month_x, format = '%d-%b-%y')
final['Month_y']= pd.to_datetime(final.Month_y, format = '%d-%b-%y')
final.dtypes
```

```python
#Calculate the city wise spend on each product on yearly basis. Also include a graphical 
representation for the same.
df_grp = final.groupby(['City',(final.Month_x.dt.year),'Product']).Amount_x.sum().reset_index()
sns.catplot(data = df_grp, x= 'Month_x', y = 'Amount_x',hue = 'Product',col='City',kind ='bar')
```

```python
#Monthly comparison of total spends, city wise
cty_spend = final.groupby(['City',(final.Month_x.dt.month)]).Amount_x.sum().reset_index()
sns.catplot(data = cty_spend,x='City' , y ='Amount_x', hue='Month_x', kind='bar')
plt.xticks(rotation =45)
```

```python
# Comparison of yearly spend on air tickets
yr_spend = final.loc[final.Type == 'AIR TICKET'].groupby((final.Month_x.dt.year)).Amount_x.sum().reset_index()
plt.pie(yr_spend.Amount_x, labels=yr_spend.Month_x, startangle=90, autopct='%.1f%%')
plt.show()
```

```python
# Comparison of monthly spend for each product (look for any seasonality that exists in terms of spend)
monthly_spe = final.groupby([(final.Month_x.dt.month),'Product']).Amount_x.sum().reset_index()
sns.catplot(data= monthly_spe, x='Month_x', y='Amount_x',hue='Product',kind='bar')
```

```python
# You need to find top 10 customers for each city in terms of their repayment amount by 
#different products and by different time periods i.e. year or month. The user should be able 
#to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the 
#function should automatically take these inputs while identifying the top 10 customers.
def f_prodct_year (product, timeperiod):
    if product.lower() =='gold' and timeperiod =='monthly':
        result = final.loc[final.Product =='Gold'].groupby([(final.Month_y.dt.month),'City','Customer']).Amount_y.sum().reset_index()
        f_res = result.sort_values(['Month_y','City','Amount_y'],ascending = [True,True,False])
    elif product.lower() =='platinum' and timeperiod =='monthly':
        result = final.loc[final.Product =='Platimum'].groupby([(final.Month_y.dt.month),'City','Customer']).Amount_y.sum().reset_index()
        f_res = result.sort_values(['Month_y','City','Amount_y'],ascending = [True,True,False])
    if product.lower() =='silver' and timeperiod=='monthly':
        result = final.loc[final.Product =='Silver'].groupby([(final.Month_y.dt.month),'City','Customer']).Amount_y.sum().reset_index()
        f_res = result.sort_values(['Month_y','City','Amount_y'],ascending = [True,True,False])
    elif product.lower() =='gold' and timeperiod =='yearly':
        result = final.loc[final.Product =='Gold'].groupby([(final.Month_y.dt.year),'City','Customer']).Amount_y.sum().reset_index()
        f_res = result.sort_values(['Month_y','City','Amount_y'],ascending = [True,True,False])
    if product.lower() =='silver' and timeperiod =='yearly':
        result =final.loc[final.Product =='Silver'].groupby([(final.Month_y.dt.year),'City','Customer']).Amount_y.sum().reset_index()
        f_res = result.sort_values(['Month_y','City','Amount_y'],ascending = [True,True,False])
    elif product.lower() =='platinum' and timeperiod =='yearly':
        result = final.loc[final.Product =='Platimum'].groupby([(final.Month_y.dt.year),'City','Customer']).Amount_y.sum().reset_index()
        f_res = result.sort_values(['Month_y','City','Amount_y'],ascending = [True,True,False])
    return f_res
```
